```python
plt.rcParams["figure.figsize"] = (5,7)
print("Plot of Class Count :: \n")
class_dict = {"Authentic" : df[df['class'] == 0]['class'].count(), 'Forged': df[df['class'] == 1]['class'].count()}

class_type = list(class_dict.keys())
class_count = list(class_dict.values())

plt.bar(class_type[0], class_count[0], color=['green'], width = 0.1)
plt.bar(class_type[1], class_count[1], color=['red'], width = 0.1)

plt.title("Class Count", fontsize=20)
plt.xlabel("Class", fontsize= 14)
plt.ylabel("Count", fontsize= 14)
plt.yticks(np.arange(0,850,50))
plt.legend(labels=["Authentic = 0", "Forged = 1"])
plt.show()


```

```python
x = df.drop(['class'], axis = 1)
x.shape
```

```python
y = df['class']
y.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(x_train, y_train)

y_pred = lr.predict(x_test)
lr.score(x_test, y_test), lr.score(x_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(classification_report(y_test, y_pred))
print(f"Accuracy= {accuracy_score(y_test, y_pred)}")
```
