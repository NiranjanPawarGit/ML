```python
x = df.drop(['animal_name', 'Class_Type', 'Class_Number'], axis=1)
xn = df.drop(['animal_name', 'Class_Type', 'Class_Number'], axis =1)
y = df['Class_Number']
x.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit_transform(xn)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
xn_train, xn_test, y_train, y_test = train_test_split(xn, y, test_size=0.2, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
```

```python
knn_n = KNeighborsClassifier(n_neighbors=5)
knn_n.fit(xn_train, y_train)
```

```python
y_pred = knn.predict(x_test)
yn_pred = knn.predict(xn_test)
```

```python
knn.score(x_test, y_test), knn.score(x_train, y_train), knn.score(xn_test, y_test), knn.score(xn_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

```

```python
print(classification_report(y_test,y_pred))
```

```python
print(classification_report(y_test,yn_pred))
```

```python
from sklearn.model_selection import cross_val_score
k_list = np.arange(1, 50, 2)
mean_scores = []
accuracy_list = []
error_rate = []

for i in k_list:
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    pred_i = knn.predict(x_test)
    score = cross_val_score(knn,x_train, y_train,cv=10)
    mean_scores.append(np.mean(score))
    error_rate.append(np.mean(pred_i != y_test))

print("Mean Scores:")
print(mean_scores)
print("Error Rate:")
print(error_rate)
```

```python
plt.plot(k_list,mean_scores, marker='o')

# Added titles and adjust dimensions
plt.title('Accuracy of Model for Varying Values of K')
plt.xlabel("Values of K")
plt.ylabel("Mean Accuracy Score")
plt.xticks(k_list)
plt.rcParams['figure.figsize'] = (12,12) 

plt.show()
```
